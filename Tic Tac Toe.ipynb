```python
from IPython.display import clear_output
import time

# function to print emoji selection screen
def print_symbols():
    print("************************ Choose your Symbol **************************")
    print(".--------------------------------------------------------------------.")
    print("|                                                                    |")
    print("| 1  ğŸ˜ | 2  ğŸ¥° | 3  ğŸ¤‘ | 4  ğŸ˜ˆ | 5  ğŸ˜œ | 6  ğŸ˜¸ | 7  ğŸ¦Š | 8  ğŸ¶  | ")
    print("|                                                                    |")
    print("'--------------------------------------------------------------------'")
    print("|                                                                    |")
    print("| 9  ğŸ± | 10 ğŸ¥ | 11 ğŸ’ | 12 ğŸ€ | 13 ğŸ” | 14 ğŸ¤– | 15 ğŸµ | 16 ğŸ¤¡  |")
    print("|                                                                    |")
    print("'--------------------------------------------------------------------'")
    

def input_validation(rang, msg, p1_char = ' '):
    choice = ' '
    while choice not in rang:
        if p1_char == ' ':
            clear_output()
            print_symbols()
            choice = input(msg)
        else :
            clear_output()
            print_symbols()
            print(f'Player1 {p1_char}')            
            choice = input(msg)
        
        if choice.isdigit():
            choice = int(choice)
           
            
    return choice


# 1. take input and return the character
def symbol_select(player_name, p1_char = ' '):
    symbols = ['ğŸ˜', 'ğŸ¥°', 'ğŸ¤‘', 'ğŸ˜ˆ', 'ğŸ˜œ', 'ğŸ˜¸', 'ğŸ¦Š', 'ğŸ¶', 'ğŸ±', 'ğŸ¥', 'ğŸ’', 'ğŸ€', 'ğŸ”', 'ğŸ¤–', 'ğŸµ', 'ğŸ¤¡']
    p_char_index = input_validation(range(1,17), player_name + ' ', p1_char) - 1
    p_char = symbols[p_char_index]
    print(f'{p_char}\n')
    
    return p_char


def print_grid():
    print('----------------------')
    print(f'|  {grid[0]}  |  {grid[1]}  |  {grid[2]}  |')
    print('----------------------')
    print(f'|  {grid[3]}  |  {grid[4]}  |  {grid[5]}  |')
    print('----------------------')
    print(f'|  {grid[6]}  |  {grid[7]}  |  {grid[8]}  |')
    print('----------------------')
    

def user_input(marking_char):

    choice = ' '
    isempty = False
    
    while choice.isdigit() == False or isempty == False:
        choice = input('Type index location [1-9] : ')

        # choice is not a number
        if choice.isdigit() == False:
            clear_output()
            print_grid()
            print(f"{marking_char}'s turn")
            print(f'Sorry, {choice} is not a number')
            
        #choice is a number
        elif choice.isdigit() == True: 

            inrange =  int(choice) in range(1, 10)
            
            if inrange:
                isempty = grid[int(choice)-1] == '  '
            if not inrange:
                clear_output()
                print_grid()
                print(f"{marking_char}'s turn")
                print(f'Sorry the number {choice} is not in range[1-9]')
            elif not isempty:
                clear_output()
                print_grid()
                print(f"{marking_char}'s turn")
                print(f'Sorry the position {choice} is already filled')
    
    return int(choice)-1


def mark_grid(grid, position, char):
    grid[position] = char
    
def change_turn(player_turn):
    return not player_turn

def set_marking_char(turn, player1char, player2char):
    if turn:
        return player1char
    else:
        return player2char
    
def full_board(grid, player1char, player2char):
    count = 0
    for i in grid:
        if i in (player1char, player2char):
            count += 1
               
    return count == len(grid)

def winning_condition(grid,c):
    for i in range(0,7,3):
        if (grid[0+i] == c)  and (grid[1+i] == c) and (grid[2+i] == c):
            return True
        
    for i in range(0,3):
        if (grid[0+i] == c)  and (grid[3+i] == c) and (grid[6+i] == c):
            return True
        
    diagonal1 = grid[0]==c and grid[4]==c and grid[8]==c 
    diagonal2 = grid[2]==c and grid[4]==c and grid[6]==c 
    return diagonal1 or diagonal2


```
